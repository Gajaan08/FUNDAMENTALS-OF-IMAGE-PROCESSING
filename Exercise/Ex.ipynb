```python
# Gaussian smoothing to filter the noise in the image
import cv2 as cv
import matplotlib.pyplot as plt
im=cv.imread('gal_gaussian.png')
im=cv.cvtColor(im,cv.COLOR_BGR2RGB)
blur=cv.blur(im,(5,5),0)
Gblur=cv.GaussianBlur(im,(5,5),0)
plt.subplots(figsize=(12, 8))
plt.subplot(131),plt.imshow(im),plt.title("Original Image"),plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(blur),plt.title("Average Smoothing Image"),plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(Gblur),plt.title("Gaussian blurring Smoothing Image"),plt.xticks([]), plt.yticks([])
plt.show()
```

Averaging
This is done by convolving an image with a normalized box filter. It simply takes the average of all the pixels under the kernel area and replaces the central element. This is done by the function cv.blur() or cv.boxFilter().

Gaussian Blurring
In this method, instead of a box filter, a Gaussian kernel is used. It is done with the function, cv.GaussianBlur(). We should specify the width and height of the kernel which should be positive and odd. We also should specify the standard deviation in the X and Y directions, sigmaX and sigmaY respectively. If only sigmaX is specified, sigmaY is taken as the same as sigmaX. If both are given as zeros, they are calculated from the kernel size. Gaussian blurring is highly effective in removing Gaussian noise from an image.

```python
# Median filtering to filter the salt and pepper noise
import cv2 as cv
import matplotlib.pyplot as plt
im=cv.imread('gal_sandp.png')
im=cv.cvtColor(im,cv.COLOR_BGR2RGB)
Mblur = cv.medianBlur(im, 5)
plt.subplots(figsize=(8, 8))
plt.subplot(121),plt.imshow(im),plt.title("Original Image"),plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(Mblur),plt.title("Median Filter Image"),plt.xticks([]), plt.yticks([])
plt.show()
```

Median Blurring
Here, the function cv.medianBlur() takes the median of all the pixels under the kernel area and the central element is replaced with this median value. This is highly effective against salt-and-pepper noise in an image. Interestingly, in the above filters, the central element is a newly calculated value which may be a pixel value in the image or a new value. But in median blurring, the central element is always replaced by some pixel value in the image. It reduces the noise effectively. Its kernel size should be a positive odd integer

```python
# Gamma Correction
import cv2 as cv
import matplotlib.pyplot as plt
img_orig=cv.imread('spider.png')
img_orig=cv.cvtColor(img_orig,cv.COLOR_BGR2RGB)
plt.subplots(figsize=(18, 8))
plot=[152,153,154,155]
gamma=[0.2,0.8,1.2,2]
plt.subplot(151),plt.imshow(img_orig),plt.title("Original Image"),plt.xticks([]), plt.yticks([])
for g in range(4) :
    table=np.array([(i/255.0)**(gamma[g])*255.0 for i in np.arange(0,256)]).astype('uint8')
    img_gamma = cv.LUT(img_orig,table)
    plt.subplot(plot[g]),plt.imshow(img_gamma),plt.title("Gamma Corrected ("+str(gamma[g])+") Image"),plt.xticks([]), plt.yticks([]) 
plt.show()
```
